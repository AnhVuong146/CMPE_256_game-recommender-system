# Board Games Preprocessing
---


**Dataset:** [Board Games Database from BoardGameGeek](https://www.kaggle.com/datasets/threnjen/board-games-database-from-boardgamegeek)

This notebook is simply to apply transforms and feature engineering on the datasets to be exported for training. If you would like to see the reasoning behind the cleaning, please checkout eda.ipynb

```python
import kagglehub
import pandas as pd
import numpy as np
```

```python
path = kagglehub.dataset_download("threnjen/board-games-database-from-boardgamegeek")

games = pd.read_csv(f"{path}/games.csv")
user_ratings = pd.read_csv(f"{path}/user_ratings.csv")
mechanics = pd.read_csv(f"{path}/mechanics.csv")
subcategories = pd.read_csv(f"{path}/subcategories.csv")
artist = pd.read_csv(f"{path}/artists_reduced.csv")
publisher = pd.read_csv(f"{path}/publishers_reduced.csv")
```

```python
# create user-item df

# map user items to id
user_to_id = {user: idx for idx, user in enumerate(user_ratings["Username"].unique())}
item_to_id = {item: idx for idx, item in enumerate(user_ratings["BGGId"].unique())}

# apply mappings
user_item_df = pd.DataFrame({
    "user": user_ratings["Username"].map(user_to_id),
    "item": user_ratings["BGGId"].map(item_to_id),
    "rating": user_ratings["Rating"]
})

# sanity check
assert user_item_df.isnull().sum().sum() == 0
```

```python
# create predicted average ratings

rating = user_ratings.groupby(by=["BGGId"], dropna=False)["Rating"].mean()
rating = pd.DataFrame(rating.reset_index())
rating = rating.drop(["BGGId"], axis=1)
rating.head()
```

```python
# drop unwanted columns and sort by id

artist = artist.sort_values(by=['BGGId'])
publisher = publisher.sort_values(by=['BGGId'])
subcat = subcategories.sort_values(by=['BGGId'])
games = games.sort_values(by=['BGGId'])

artist = artist.drop(["BGGId"], axis=1)
publisher = publisher.drop(["BGGId"], axis=1)
subcat = subcat.drop(["BGGId"], axis=1)
games = games.drop(["BGGId", "Family", "ComAgeRec", "LanguageEase", "Description", "ImagePath", "Name",
                    "Rank:boardgame", "Rank:strategygames", "Rank:abstracts",
                    "Rank:familygames", "Rank:thematic", "Rank:cgs", "Rank:wargames",
                    "Rank:partygames", "Rank:childrensgames",
                    "Cat:Thematic", "Cat:Strategy", "Cat:War", "Cat:Family",
                    "Cat:CGS", "Cat:Abstract", "Cat:Party", "Cat:Childrens"], axis=1)
```

```python
# combine metadata into one set

df = pd.concat([games, artist, publisher, subcat], axis=1)
df.head()
```

Saving locally or through vm

```python
%store rating df user_item_df
```

```python
df.to_csv("df.csv")
rating.to_csv("ratings.csv")
user_item_df.to_csv("user_item.csv")
```
